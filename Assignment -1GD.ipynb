```python
###  dataset creation 
import numpy as np
import pandas as pd
N = 100
X  = np.linspace(0,10,N)
Y =  2*X+4+np.random.normal(0,1,100)       ### y  = m1*X+c+N(0,1)
df = pd.DataFrame({'X': X,'Y':Y})
df.head()
```

```python
########### visualize the dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.scatter(X,Y)

plt.show()
```

```python
### compute_loss
```

```python
def compute_loss(y_true,y_pred):
    return np.mean((y_true-y_pred)**2)
```

```python
####  Gradient descent 
m1 = np.random.randn()
```

```python
########### graient funtion
loss_lst =[]
m_values =[]
def fun(X,Y,lr =0.01,epochs=1000):
    global m1
    N = len(X)
    for epoch in range(epochs):
        y_pred = m1*X+4
        loss = compute_loss(Y,y_pred)
        m_values.append(m1)
        loss_lst.append(loss)
        grad  = (-2/N)*np.sum(X*(Y-y_pred))
        m1 = m1-lr*grad
        if epoch%100==0:
            print(loss)
    return m1
m1 = fun(X,Y)
print(f'M1 value{m1}')
```

```python
############## GD  graphical view 
```

```python
plt.plot(m_values,loss_lst)
plt.show()
```

```python
###  dataset creation 
import numpy as np
import pandas as pd
N = 100
X  = np.linspace(0,10,N)
Y =  2*X+4+np.random.normal(0,1,100)       ### y  = m1*X+c+N(0,1)
df = pd.DataFrame({'X': X,'Y':Y})
df.head()
```

```python

def compute_loss(y_true,y_pred):
    return np.mean((y_true-y_pred)**2)
m1_val  = np.linspace(-18,23,1000)
loss_lst  = []
best_m1 = None
min_loss = float('inf')
for m1 in m1_val:
    y_pred  = m1*X+4
    loss = compute_loss(Y,y_pred)
    loss_lst.append(loss)
    if loss<min_loss:
        min_loss = loss
        best_m1 = m1
        
print(f'm1_values{best_m1}')
```

```python
plt.plot(m1_val,loss_lst,color='b')
plt.axvline(x=best_m1 , linestyle ='--' , color='r')
plt.xlabel("m1 values")
plt.ylabel("loss")
plt.show()
```
