```python
import numpy as np
import matplotlib.pyplot as plt

## class HeartShape
class HeartShape:
    def __init__(self, x_lim=(-2, 2), y_lim=(-2, 2), number=1000):
        self.x_lim = x_lim     ### set x limit [ start , end]
        self.y_lim = y_lim     ### set y limit [ start , end]
        self.number = number   ### set the number of points for smoothness

    
        self.x, self.y = np.meshgrid(
            np.linspace(self.x_lim[0], self.x_lim[1], self.number),
            np.linspace(self.y_lim[0], self.y_lim[1], self.number)
        )
     #####  heart function     
    def heart_function(self, x, y):
        return (x**2 + y**2 - 1)**3 - x**2 * y**3
    ##### plot the  graph 
    def plot_heart(self):
        z = self.heart_function(self.x, self.y)
        plt.contour(self.x, self.y, z, levels=[0], colors='r')
        plt.title("Heart Shape Diagram")
        plt.show()

#To call the class 
heart = HeartShape()
# to call the function
heart.plot_heart()
```
